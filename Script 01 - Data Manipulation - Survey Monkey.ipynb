```python
cols_to_drop= ['Start Date',
                 'End Date',
                 'Email Address',
                 'First Name',
                 'Last Name',
                 'Custom Data 1']

dataset = dataset.drop(columns=cols_to_drop)

dataset.info()
```

```python
list(dataset.columns)[:8]
```

```python
id_vars = list(dataset.columns)[:8]
value_vars = list(dataset.columns)[8:]
```

Melt doc -> https://pandas.pydata.org/pandas-docs/version/1.0.0/reference/api/pandas.DataFrame.melt.html

```python
# melt is used to pivot dataframe 

# Assigning each Q&A to the demographic columns 

test_dataset= dataset.melt(id_vars=id_vars,value_vars=value_vars,var_name="Question+subquestion",value_name= "Answers")

test_dataset
```

```python
dataframe = test_dataset
```

**Adding the original Questions without the sub questions**

```python
questions = pd.read_excel(pwd+"/Edited Survey Monkey Output.xlsx",sheet_name="Questions")
questions
```

```python
questions_backup = questions.copy()
```

```python
questions.columns
```

```python
cols_to_drop = ['Raw Question', 'Raw Subquestion', 'Subquestions', 'Unnamed: 5']

questions.drop(columns=cols_to_drop , inplace = True)

questions
```

**Now we'll join the two datasets**

pd.merge doc -> https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html

```python
# When using joint, make sure the original col number stays the same. 
# otherwise indicates duplicate instances got mapped and created redundent data

df_joint = pd.merge(left=dataframe,right=questions,how="left" , left_on="Question+subquestion", right_on= "Question+subquestion")

print("OG len : " ,len(dataframe))
print("Marged len: ", len(df_joint))

df_joint_cpy=df_joint.copy()
df_joint
```

### 3 - how many people answered the same question 

```python
# fishing out all the answers that aren't null  

respondents = df_joint[df_joint["Answers"].notnull()]

# Grouping NonNull answers by Questions, using Unique Respondent Id to count unique respondenet number. 

respondents = respondents.groupby("Question")["Respondent ID"].nunique().reset_index()

respondents.rename(columns={"Respondent ID":"Respondents"},inplace=True)

respondents
```

```python
df_joint_2 = pd.merge(left=df_joint,right=respondents,how="left" , left_on="Question", right_on= "Question")

print("OG len : " ,len(df_joint))
print("Marged len: ", len(df_joint_2))

df_joint_2_cpy=df_joint_2.copy()
df_joint_2
```

### 4 - How many people gave the same answer 

```python

# Based on Questions and Sub Questions, group by answers 
# Count number of respondents by answers 

# fishing out all the answers, including the null ones

same_ans = df_joint_2#[df_joint["Answers"].notna()]


same_ans = same_ans.groupby(["Question+subquestion","Answers"])["Respondent ID"].nunique().reset_index()

same_ans.rename(columns={"Respondent ID":"Same Answer Count"},inplace=True)

same_ans
```

```python
df_joint_3 = pd.merge(left=df_joint_2,right=same_ans,how="left" , left_on=["Question+subquestion","Answers"], right_on= ["Question+subquestion","Answers"])

df_joint_3["Same Answer Count"].fillna(0,inplace=True)

print("OG len : " ,len(df_joint_2))
print("Marged len: ", len(df_joint_3))

df_joint_3_cpy=df_joint_3.copy()
df_joint_3
```

### 5 - Touch up the naming convention 

```python
output =  df_joint_3.copy()
output.rename(columns={"Identify which division you work in. - Response":"Division Primary", 
                       "Identify which division you work in. - Other (please specify)":"Division Secondery" ,
                       "Which of the following best describes your position level? - Response":"Position", 
                       "Which generation are you apart of? - Response":"Generation" ,
                       "Please select the gender in which you identify. - Response":"Gender",
                       "Which duration range best aligns with your tenure at your company? - Response" :"Tenure", 
                       "Which of the following best describes your employment type? - Response":"Employment type",
                       "Question+subquestion":"Question and Sub-Question"}, inplace=True)

output.head()
```
